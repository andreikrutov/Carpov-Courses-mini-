```python
df = pd.read_csv('/content/drive/My Drive/Data for Collab/Carpov_Courses/dop_6/gff.tsv', sep='\t')
```

```python
def extracting(attr):
"""
Находим индекс Parent отсчитаываем от него количество символов, соответствующее Parent= 
и возвращаем все, что стоит после
"""
  ind = attr.index('Parent')
  ind = ind + 7

  return attr[ind:]
```

```python
df['extr_attributes'] = df['attributes'].apply(extracting)
```

```python
df.groupby('extr_attributes').agg({'phase':'count'}).idxmax()
```

Какое самое частое значение в колонке type?

```python
df.groupby('type', as_index=False).agg({'phase':'count'})
```
